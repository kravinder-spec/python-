# ***if-elif statements***

```python
marks=input("enter student marks")
marks=int(marks)
if marks>75:
  print("student pass in the subject with A grade")
elif marks>60:
  print("student pass in the subject with B grade")
elif marks>45:
  print("student pass in the subject with C grade")
else:
  print("student fail in the subject")
```
